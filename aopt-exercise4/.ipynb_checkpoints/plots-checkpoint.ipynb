```python
def f(x, y):
    return x*x-2*y*y

def fi1(x,y):
    return np.less(np.power(x+4,2)-2-y,0)
x = np.linspace(-5, 5, 401)
y = np.linspace(-5, 5, 401)

def fe1(x,y):
    return np.less(np.abs(x-y+4),1e-8)

def fi2(x,y):
    return np.less(-x-10,0)

X, Y = np.meshgrid(x, y)
Z = f(X,Y)

mask1 = fi1(X,Y)

mask2 = fe1(X,Y)

mask3 = fi2(X,Y)

Z = np.take(Z,(mask1).nonzero())
X = np.take(Y,(mask1).nonzero())
Y = np.take(Y,(mask1).nonzero())


%matplotlib widget
fig = plt.figure()
ax = plt.axes(projection='3d')
#surf= ax.plot_surface(X*mask2*mask1*mask3, Y*mask2*mask1*mask3, Z*mask2*mask1*mask3, cmap= 'cool')
#fig.colorbar(surf, shrink=0.5, aspect=5)
#ax.contour3D(X, Y, Z, 50, rstride=1, cstride=1)

surf = ax.plot_surface(X, Y, Z, cmap='bwr')
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

```
