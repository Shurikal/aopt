```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return x*x-2*y*y


X, Y = np.meshgrid(x, y)
Z = f(X,Y)


#%matplotlib widget
fig = plt.figure()
ax = plt.axes(projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='bwr')
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

```

```python
def f(x, y):
    return x*x-2*y*y

def fi1(x,y):
    return np.less(np.power(x+4,2)-2-y,0)
x = np.linspace(-5, 5, 401)
y = np.linspace(-5, 5, 401)

def fe1(x,y):
    return np.less(np.abs(x-y+4),1e-8)

def fi2(x,y):
    return np.less(-x-10,0)

X, Y = np.meshgrid(x, y)
Z = f(X,Y)

mask1 = fi1(X,Y)

mask2 = fe1(X,Y)

mask3 = fi2(X,Y)


%matplotlib widget
fig = plt.figure()
ax = plt.axes()

surf = ax.contourf(X, Y, Z*mask3*mask1*mask2,40,cmap="turbo_r")
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_xlabel('x')
ax.set_ylabel('y')
#ax.set_zlabel('z');


Z = Z*mask1*mask2*mask3

print(Z)



minElement = np.amin(Z)

print('Minimum element from Numpy Array : ', minElement)

result = np.where(Z == np.amin(Z))
print('Returned tuple of arrays :', result)

i = result[0][0]
j = result[1][0]
print(X[i,j]," , ", Y[i,j])
```
